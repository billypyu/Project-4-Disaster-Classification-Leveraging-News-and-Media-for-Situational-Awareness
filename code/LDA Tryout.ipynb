```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, ENGLISH_STOP_WORDS
from sklearn.decomposition import LatentDirichletAllocation

from gensim import corpora, models
import pyLDAvis.gensim

pyLDAvis.enable_notebook()

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

np.random.seed(42)
```

```python
df = pd.read_csv('../data/e_all_keyword_list.csv')
```

```python
title_df = df[['title']]
```

```python
title_df.head()
```

### Step 1: CountVectorize Titles 

```python
cv = CountVectorizer(ngram_range = (1, 1), 
                     stop_words = 'english', 
                     min_df = 2)
cv_title = cv.fit_transform(title_df['title'])
```

```python
cv_title_df = pd.DataFrame(cv_title.toarray(), columns = cv.get_feature_names())
```

```python
cv_title_df.head()
```

```python
texts = [cv_title_df.columns[cv_title_df.loc[index,:].nonzero()] for index in cv_title_df.index]
```

```python
texts[:5]
```

```python
dictionary = corpora.Dictionary(texts)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
ldamodel = models.ldamodel.LdaModel(corpus,                     # pass in our corpus
                                    id2word = dictionary,       # matches each word to its "number" or "spot" in the dictionary
                                    num_topics = 5,             # number of topics T to find
                                    passes = 5,                 # number of passes through corpus; similar to number of epochs
                                    minimum_probability = 0.01)
```

```python
pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
```
